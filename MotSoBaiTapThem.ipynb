```python
# Bài 28: Tính nCk (Dùng def)

def factorial(n):
    if n == 0:
        return 1
    elif n < 0:
        return False
    else:
        res = 1
        for i in range(1, n+1):
            res *= i
        return res
        
def tohop(n, k):
    if n < k or n < 0 or k < 0:
        return False
    else:
        return factorial(n) // (factorial(k) * factorial(n-k))
        
try:
    n, k = map(int, input("Nhập n, k: ").split())
    print(tohop(n, k))
except ValueError:
    print("!!! Đầu vào không hợp lệ")
```

```python
# Bài 29: Lập tam giác Pascal, bằng nCk (Dùng def)

def factorial(n):
    if n == 0:
        return 1
    elif n < 0:
        return False
    else:
        res = 1
        for i in range(1, n+1):
            res *= i
        return res
        
def combination(n, k):
    return factorial(n) // (factorial(k) * factorial(n-k))

try:
    n = int(input('Nhập số hàng: '))
    if n >= 1:
        for i in range(n):
            # print(' ' * (n-i-1), end='')  # --Làm tam giác
            for j in range(i+1):
                print(combination(i, j), end=' ')
            print()
    else:
        print('!! Nhập số nguyên dương')
except ValueError:
    print('!!! Đầu vào không hợp lệ')
```

```python
# Bài 30: Viết def tính diện tích hình và có MENU

from math import sqrt

def square():
    try:
        a = float(input('Nhập cạnh hình vuông:'))
        if a <= 0:
            return False
        else:
            return f"Diện tích = {a**2}"
    except ValueError:
        return '!! Đầu vào không hợp lệ'

def circle():
    try:
        r = float(input('Nhập bán kính hình tròn:'))
        if r <= 0:
            return False
        else:
            return f"Diện tích = {3.14 * (r**2)}"
    except ValueError:
        return '!! Đầu vào không hợp lệ'

def triangle():
    try:
        a, b, c = map(float, input('Nhập ba cạnh của tam giác:').split())
        x, y, z = sorted([a,b,c])
        if (x > 0) and (x + y > z):
            p = (x+y+z) / 2
            return f"Diện tích = {sqrt(p*(p-x)*(p-y)*(p-z))}"
        else: 
            return False
    except ValueError:
        return '!! Đầu vào không hợp lệ'

def rectangle():
    try:
        a, b = map(float, input('Nhập độ dài hai cạnh của hình chữ nhật:').split())
        if (a > 0) and (b > 0):
            return f"Diện tích = {a*b}"
        else:
            return False
    except:
        return '!! Đầu vào không hợp lệ'

print('----------MENU----------')
print('0. Không làm gì hết và trở về màn hình soạn thảo')
print('1. Tính diện tích hình vuông')
print('2. Tính diện tích hình tròn')
print('3. Tính diện tích tam giác')
print('4. Tính diện tích hình chữ nhật')
print('-' * 30)

try:
    num = int(input('Chọn số: '))
    
    while not 0 <= num <= 4:
        print('-' * 30)
        print('!!! Chỉ chọn trong menu')
        num = int(input('Chọn số: '))                
    
    match num:
        case 0:
            print('---END---')
        case 1:
            print(square())
        case 2:
            print(circle())
        case 3:
            print(triangle())
        case 4:
            print(rectangle())
            
except ValueError:
    print('Đầu vào không hợp lệ')

```

```python
# Bài 32: Tìm số đảo có số nguyên dương

def reverse(n):
    if n <= 0:
        return False
    else:
        rev = 0
        
        while n > 0:
            i = n % 10          # Lấy hàng đơn vị
            rev = rev * 10 + i  # rev * 10 + i
            n //= 10            # Bỏ hàng đơn vị
            
        return rev
        
try:
    n = int(input('Nhập n nguyên dương:'))
    print(f"Số đảo của {n} là {reverse(n)}")
except ValueError:
    print('Đầu vào không hợp lệ')
```

```python
# Bài 34: Tính tổng x^i với i chạy từ 1 -> n (Dùng def)

def sigma_sum(x, n):
    if n > 0:
        s = 0
        temp = 1
        for i in range(1, n+1):
            temp *= x
            s += temp
        return s
    else:
        return '!!! Nhập n nguyên dương'

try:
    x = float(input('Nhập x:'))
    n = int(input('Nhập n:'))
    print(sigma_sum(x, n))
except ValueError:
    print('!! Đầu vào không hợp lệ')
```

```python
# Bài 35: Tính tổng (1+x^i) / (1+i)! với i chạy từ 0 -> n (Dùng def)

def sigma(x,n):
    if x == 0 or n < 0:
        return False
    elif n == 0:
        return 2
    else:
        temp_up = temp_down = 1
        s = 2
        for i in range(1,n+1):
            temp_up *= x
            temp_down *= (i+1)
            s += (temp_up + 1) / temp_down
        return s

try:
    n = int(input('Nhập n:'))
    x = float(input('Nhập x:'))
    print(sigma(x,n))
except ValueError:
    print('!! Đầu vào không hợp lệ')
```

```python
# Bài 39: Nhập một ma trận, in ra ma trận chuyển vị

def display(A):
    for i in A:
        for j in i:
            print(f"{j:g}", end=' ')
        print()

try:
    m, n = map(int, input('Nhập cỡ của ma trận (Hàng _ Cột):').split())
    if m > 0 and n > 0:
        M = []
        
        # Tìm ma trận chuyển vị của M luôn
        print('-' * 30)
        A = [[0 for _ in range(m)] for _ in range(n)] # Tạo ma trận cỡ n x m (Chuyển vị)
        j = 0
        for _ in range(m):
            row = list(map(float, input().strip().split()))
            M.append(row)
            # -- Chuyển luôn hàng thành cột
            for i in range(n):
                A[i][j] = row[i]
            j += 1
        # --> Lúc này A là ma trận chuyển vị của M
            
        print('-' * 30)
        
        # Ma trận chuyển vị:
        print('Ma trận chuyển vị:')
        display(A)

        print('-' * 30)

        # In ra dạng chuyển vị:
        print("'In ra dạng chuyển vị'")
        for i in range(n):
            for j in range(m):
                print(f"{M[j][i]:g}", end=' ')
            print()
            
except ValueError:
    print('!! Đầu vào không hợp lệ')
            
```

```python
# Bài 53: Viết chương trình nhập vào ma trận vuông cấp n:
# a) In ra các phần tử trên đường chéo chính
# b) In ra các phần tử trên đường chéo phụ

try:
    n = int(input('Nhập cấp của ma trận vuông:'))
    if n > 0:
        print('-' * 30)
        print('Nhập ma trận')
        flag = True          # -- Đánh dấu nhập ma trận
        main_diagonal = []   # -- Đường chéo chính
        sub_diagonal = []    # -- Đường chéo phụ
        for i in range(n):
            row = list(map(float, input().strip().split()))
            if len(row) != n:
                print('!! Số lượng phần tử không đúng')
                flag = False
                break
            else:
                main_diagonal.append(f"{row[i]:g}")
                sub_diagonal.insert(0, f"{row[n-i-1]:g}")
        if flag:
            print('-' * 30)
            print('Đường chéo chính:', *main_diagonal)
            print('Đường chéo phụ:', *sub_diagonal)
        
    else:
        print('!! Nhập n nguyên dương')
except ValueError:
    print('!! Đầu vào không hợp lệ')
```

```python
# Bài 54: Viết chương trình nhập vào ma trận vuông cấp n.
# Kiểm tra ma trận đó có đối xứng hay không

try:
    n = int(input('Nhập cấp của ma trận vuông:'))
    if n > 0:
        print('-' * 30)
        print('Nhập ma trận')
        flag = True          # -- Đánh dấu nhập ma trận
        matrix = []          # -- Khởi tạo ma trận
        for i in range(n):
            row = list(map(float, input().strip().split()))
            if len(row) != n:
                print('-' * 30)
                print('!! Số lượng phần tử không đúng')
                flag = False
                break
            else:
                matrix.append(row)
                
        if flag:
            print('-' * 30)
            symmetric = True   # -- Giả sử đang là ma trận đối xứng
            for i in range(n):
                for j in range(i+1, n):
                    if matrix[i][j] != matrix[j][i]:
                        symmetric = False
                        print('Không phải ma trận đối xứng')
                        break
                if symmetric == False:
                    break

            if symmetric:
                print('Đây là ma trận đối xứng')
        
    else:
        print('!! Nhập n nguyên dương')
except ValueError:
    print('!! Đầu vào không hợp lệ')
```

```python
# Bài 55: Viết chương trình nhập vào hai ma trận vuông cấp n
# a) Tính tổng hai ma trận
# b) Tính tích hai ma trận
# -- Hàm zip() là lặp song song

def display(A):
    for i in A:
        for j in i:
            print(f"{j:g}", end=' ')
        print()
        
def input_matrix(m,n):
    M = []
    for i in range(m):
        row = list(map(float, input().strip().split()))
        if len(row) != n:
            print('-' * 30)
            print('!! Số lượng phần tử không chính xác')
            return []
        else:
            M.append(row)
    return M
                     
def plusmatrix(A, B):
    sAB = [[a+b for a,b in zip(rowA, rowB)] for rowA, rowB in zip(A, B)]
    return sAB

def timesmatrix(A, B):
    BT = list(zip(*B))
    AB = [[sum(a*b for a, b in zip(rowA, columnB)) for columnB in BT] for rowA in A]
    return AB

def plus_times_matrix_tradi(A, B):
    sAB = [[0 for _ in range(n)] for _ in range(n)]   # -- Ma trận tổng
    AB = [[0 for _ in range(n)] for _ in range(n)]    # -- Ma trận tích
    for i in range(n):
        for j in range(n):
            sAB[i][j] += A[i][j] + B[i][j]      # -- Tính tổng
            for k in range(n):
                AB[i][j] += A[i][k] * B[k][j]   # -- Tính tích
    return sAB, AB

try:
    n = int(input('Nhập cấp của ma trận vuông:'))
    if n > 0:  
        
        # -- Nhập ma trận A
        print('-' * 30)
        print('Nhập ma trận A:')
        A = input_matrix(n,n)

        if A:   # -- Nếu nhập đúng ma trận A thì nhập tiếp ma trận B
            # -- Nhập ma trận B
            print('-' * 30)
            print('Nhập ma trận B:')
            B = input_matrix(n,n)

        if A and B:
            print('-' * 30)
            print('TỔNG HAI MA TRẬN:')
            display(plusmatrix(A, B))
            print('TÍCH HAI MA TRẬN:')
            display(timesmatrix(A, B))
        
    else:
        print('!! Nhập n nguyên dương')
except ValueError:
    print('!! Đầu vào không hợp lệ')
```

```python
# Bài 56: Viết chương trình đếm số từ (word) trong một chuỗi

def word_count(s):
    words = list(s.strip().split())
    return len(words)

def word_count_tradi(s):
    # -- Nếu gặp kí tự thường thì flag = True
    # -- Nếu gặp kí tự trống thì check flag
    # -- Nếu biến flag chuyển từ True sang False thì count++
    count = 0
    s += ' '
    flag = False
    for i in s:
        if not i.isalpha():
            if flag == True:
                flag = False
                count += 1
        else:
            flag = True
            
    return count

try:
    s = input('Nhập chuỗi:')
    print('-' * 30)
    print('Số từ trong chuỗi:', word_count_tradi(s))
except ValueError:
    print('!! Đầu vào không hợp lệ')
```

```python
# Bài 57: Trộn hai mảng đã được sắp xếp theo thứ tự tăng dần lại thành một mảng mới
# mà vẫn đảm bảo thứ tự của nó. Không dùng bất cứ thuật toán sắp xếp nào

try:
    A = list(map(int, input('Nhập mảng A:').strip().split()))
    B = list(map(int, input('Nhập mảng B:').strip().split()))
    if (not A and B) or (not B and A):   # -- Nếu A rỗng, B đầy hoặc B rỗng, A đầy
        C = A + B
        print('-- KẾT QUẢ --')
        print(*C)
    elif A and B:    # -- Nếu cả hai mảng khác rỗng
        C = []
        a = b = 0
        while len(C) < len(A) + len(B):
            if A[a] <= B[b]:
                C.append(A[a])
                if a < len(A) - 1:   # -- Nếu a chưa chạy đến cuối thì a++
                    a += 1
                else:                # -- Khi a đến cuối list thì cộng phần còn lại của B
                    C = C + B[b::]
            else:
                C.append(B[b])       # -- Nếu b chưa chạy đến cuối thì b++
                if b < len(B) - 1: 
                    b += 1
                else:                # -- Khi b đến cuối list thì cộng phần còn lại của A
                    C = C + A[a::]
        print('-- KẾT QUẢ --')
        print(*C)
    else:
        print('-- MẢNG RỖNG --')
except ValueError:
    print('!! Đầu vào không hợp lệ')
```

```python
# Bài 2: Viết hàm cân bằng mảng.
# Làm thế nào cho số phần tử chẵn bằng số phần tử lẻ.
# Nếu thiếu bên lẻ thì cho thêm các phần tử thiếu bằng số lẻ lớn nhất trong mảng (nếu mảng không có lẻ thì thêm 1).
# Nếu thiếu bên chẵn thì cho thêm các phần tử bằng số chẵn lớn nhất trong mảng (nếu mảng không có chẵn thì thêm 2).

def balance_even_odd(a):
    list_odd = []    # -- Mảng chứa các số lẻ
    list_even = []   # -- Mảng chứa các số chẵn

    # -- Lọc chẵn, lẻ
    for i in a:
        if i % 2 == 0:
            list_even.append(i)
        else:
            list_odd.append(i)

    distance = abs(len(list_odd) - len(list_even))     # -- Độ chênh lệch số phần tử, nếu > 0 thì odd > even và ngược lại
    if distance != 0:      # -- Chẵn != Lẻ
        
        if distance == len(a):                 # -- Tức là mảng a chứa full chẵn hoặc full lẻ
            bonus = (abs(a[0]) % 2) + 1     # -- Thêm (0+1) nếu là mảng chẵn, thêm (1+1) nếu là mảng lẻ
            
        else:                                  # -- Mảng a có chẵn có lẻ
            even_odd = [list_even, list_odd]                      # -- even_odd chứa hai mảng đã lọc
            min_type = min( even_odd, key = lambda x : len(x) )   # -- Mảng ngắn hơn
            index_min_type = even_odd.index(min_type)             # -- Vị trí mảng ngắn hơn
            bonus = max(even_odd[index_min_type])
            
        a.extend( [bonus] * distance )
        
    return a

def balance_even_odd_tradi(a):
    count_odd = count_even = 0
    max_odd = 1
    max_even = 2
    flag_odd = flag_even = False

    # -- Lọc chẵn, lẻ
    # -- Tìm max chẵn và max lẻ
    for i in a:
        if i % 2 == 0:
            count_even += 1
            if flag_even == False:
                max_even = i
                flag_even = True
            else:
                max_even = max(max_even, i)
        else:
            count_odd += 1
            if flag_odd == False:
                max_odd = i
                flag_odd = True
            else:
                max_odd = max(max_odd, i)

    distance = count_even - count_odd
    if distance > 0:    # -- Chẵn nhiều hơn lẻ
        a.extend([max_even] * distance)
    elif distance < 0:  # -- Lẻ nhiều hơn chẵn
        a.extend([max_odd] * (-distance))

    return a

try:
    a = list(map(int, input('Nhập mảng:').strip().split()))
    print(balance_even_odd(a))
    print(balance_even_odd_tradi(a))
except ValueError:
    print('!! Đầu vào không hợp lệ')
```
