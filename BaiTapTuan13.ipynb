```python
# W11A1

from pathlib import Path
WORKDIR = Path(".")  # Đường dẫn

name = "W11A1.txt"  # Tên file

def reverse(name):
    myfile = WORKDIR/name  # Đường dẫn tới file
    with open(myfile, "r") as f:
        n = int(f.read().strip())
        
    n_rev = 0
    while n:
        n_rev = n_rev * 10 + n % 10
        n //= 10
    return n_rev

print(reverse(name))
```

```python
# W11A2

from pathlib import Path
WORKDIR = Path(".")
name = "W11A1.txt"

myfile = WORKDIR/name

def canWinNim(myfile):
    with open(myfile, "r") as f:
        n = int(f.read().strip())
    if n % 4 == 0:
        return False
    return True
    
if canWinNim(myfile):
    print('Toi thang')
else:
    print('Toi thua')
```

```python
# W11A3:

from pathlib import Path
WORKDIR = Path(".")
name = "W11A1.txt"

def housesOfHogwarts(path: str):
    myfile = WORKDIR / path
    with open(myfile, "r") as f:
        n = int(f.readline().strip())
        for _ in range(n):
            print(f.readline().strip())

housesOfHogwarts(name)
```

```python
# W11A4:

from pathlib import Path
WORKDIR = Path(".")
path1 = "W11A1.txt"
path2 = "W11A2.txt"

def normalize(s: str) -> str:
    return "".join(ch.lower() for ch in s if ch.isalpha())

def areAnagrams(path1: str, path2: str) -> bool:
    file1 = WORKDIR / path1
    file2 = WORKDIR / path2
    with open(file1, "r") as f1:
        s1 = normalize(f1.read())  
    with open(file2, "r") as f2:
        s2 = normalize(f2.read()) 
    return sorted(s1) == sorted(s2)

print(areAnagrams(path1, path2))
```

```python
# W11A5:

from pathlib import Path
WORKDIR = Path(".")
path = "W11A1.txt"

def count(path: str, character:str):
    myfile = WORKDIR/path
    with open(myfile, "r") as f:
        s = f.read().strip()
    return sum(1 for ch in s if ch == character)

k = input('Nhập kí tự cần đếm:')
print(count(path, k))
```

```python
# W11A6:

from pathlib import Path
WORKDIR = Path(".")
path = "W11A1.txt"

def countCharacter(path: str, letter: str) -> int:
    myfile = WORKDIR / path
    with open(myfile, "r") as f:
        text = f.read().lower()
    return text.count(letter.lower())

k = input('Nhập kí tự cần đếm:')
print(countCharacter(path, k))
```
