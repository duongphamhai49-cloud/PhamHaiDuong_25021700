```python
# W5A1: Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số

def max_of_two(a,b):
    '''
    Returns the max of two numbers 

    Parameters:
        a (int): The first number
        b (int): The second number

    Returns:
        int: the max of two numbers
    '''
    
    if a >= b:
        return a
    else:
        return b
```

```python
# W5A3: Viết hàm kiểm tra một số có phải số nguyên tố hay không

def is_prime(n):
    '''
    Check whether given integer n is prime number.

    Parameters:
        n (int): The integer to be checked for whether it is prime

    Returns:
        bool: True if n is prime, False otherwise
    '''
    if n < 2:
        return False 
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
    
```

```python
# W5A4: Viết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lí trên dữ liệu số nguyên

def is_perfect(n):
    '''
    Check whether a given integer is a perfect number.

    Parameters:
        n (int): The integer to be checked for whether it is perfect number.

    Returns:
        bool: True if n is a perfect number, False otherwise.
    '''
    
    if n < 6:
        return False
    s = 1
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            s += i
            if i != (n//i):
                s += n // i
    return s == n 

```

```python
# W5A5: Viết hàm đưa vào một list số nguyên và một số nguyên dương k.
# Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list
# Nếu không có thì trả về -1

def find_index(a, k):
    '''
    Find the first index of k in a list of integers.

    Parameters:
        a (list): A List of integers.
        k (int): The integer to be searched for

    Returns:
        (int): The index of the first occurrence of k.
               Returns -1 if k does not appear in the list
    '''

    for i in range(len(a)):
        if a[i] == k:
            return i
    return -1

```

```python
# W5A6: Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n

def factorial(n):
    '''
    Return the factorial of a non-negative integer n

    Parameters:
        n (int): A non-negative number to calculate the factorial

    Returns:
        int: The factorial of n
    '''
    
    if n < 0:
        raise ValueError("n must be non-negative")
    
    fac = 1
    for i in range(1, n + 1):
        fac *= i
    return fac
```

```python
# W5A7: Viết hàm giả lập một máy tính bỏ túi
# Đầu vào: Hai số thực và phép toán cần thực hiện (+ - * /)
# Đầu ra: Kết quả phép toán làm tròn đến số thập phân thứ 2

def calculation(a, b, ope):
    '''
    Perform a calculation with two numbers and a given operator.

    Parameters:
        a (int or float): The first number.
        b (int or float): The second number.
        ope (str): The operator, one of (+ - * /).

    Returns:
        float: The result of the calculation, rounded to 2 decimal places.

    Raises:
        ValueError: If the operator is invalid of division by zero is attempted.
    '''

    match ope:
        case '+':
            return round(a + b, 2)
        case '-':
            return round(a - b, 2)
        case '*':
            return round(a * b, 2)
        case '/':
            if b == 0:
                raise ValueError("Cannot divide by zero")
            return round(a / b, 2)
        case _:
            raise ValueError(f"Invalid operator: {ope}")
```

```python
# W5A8: Tìm khoảng cách Hamming

def hamming_distance(x, y):
    '''
    Calculate the Hamming distance between two integers.

    Parameters:
        x (int): The first integer.
        y (int): The second integer.

    Returns:
        int: The hamming distance of two integers.
        (Number of positions at which the corresponding bits are different)
    '''
    
    xor = x ^ y #-- Phép XOR (Chỉ một): Khác nhau thì trả về 1, giống nhau thì trả về 0
    count = 0
    #-- Đếm dạng nhị phân của xor có bao nhiêu số 1 (chính là khoảng cách hamming)
    #-- Đếm bằng cách chia 2 (Như cách đổi thập phân sang nhị phân)
    while xor != 0:
        if xor % 2 == 1:
            count += 1
        xor //= 2

    return count

    # Ngắn hơn
    # return bin(x^y).count('1')
```

```python
# W5A9: Viết hàm trả về tổng các chữ số của số nguyên dương n

def sum_of_digits(n):
    '''
    Calculate the sum of the digits of a positive integer.

    Parameters:
        n (int): A positive integer whose digits will be summed.

    Returns:
        int: The sum of all digits of n.

    Raises:
        ValueError: If n is not a positive integer.
    '''

    if n<= 0:
        raise ValueError("n must be positive integer")
        
    total = 0
    while n != 0:
        total += n % 10
        n //= 10
    return total
```
