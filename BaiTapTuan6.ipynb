```python
# W6A1: Nhập vào một dòng gồm các số nguyên. 
# Hãy loại bỏ phần tử trùng nhưng giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả
# Yêu cầu in ra đúng dạng list Python, ví dụ [1,2,3]

lst = list(map(int, input().strip().split()))
s = set()  # -- Tạo set()
result = []
for x in lst:
    if x not in s:
        result.append(x)
        s.add(x)
print(result)
```

```python
# W6A2: Nhập một dòng số nguyên. 
# In ra list tổng tích lũy (phần tử i là tổng tử đầu tới vị trí i), giữ nguyên thứ tự

lst = list(map(int, input().strip().split()))
s = 0
result = []
for x in lst:
    s += x
    result.append(s)
print(result)
```

```python
# W6A3: Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế).
# In ra tuple sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp xếp)

tpl = tuple(map(int, input().strip().split()))
k = int(input())
k = k % len(tpl)
result = tpl[k:] + tpl[:k]
print(tuple(result))
```

```python
# W6A4: Nhập một dòng gồm các key:value cách nhau bởi dấu cách.
# Tạo dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện.
# Nếu key lặp lại, thêm vào list của key đó.

lst = list(map(str, input().strip().split()))
d = {}
for i in range(0,len(lst),2):
    key = lst[i]
    value = lst[i+1]
    if key not in d:
        d[key] = [value]
    else:
        d[key].append(value)

print(d)
```

```python
# W6A5: Nhập một list số nguyên
# In ra dictionary {'positives':... , 'negatives':..., 'zeros':...}
# (đếm >0, <0, =0)

lst = list(map(int, input().strip().split()))
d = {
    'positives': 0,
    'negatives': 0,
    'zeros': 0,
}
for x in lst:
    if x > 0:
        d['positives'] += 1
    elif x < 0:
        d['negatives'] += 1
    else:
        d['zeros'] += 1

print(d)
```

```python
# W6A8: Viết chương trình nhập vào một list các chuỗi.
# Sau đó đếm số lần xuất hiện của từng chuỗi và lưu vào dictionary

lst = list(map(str, input().strip().split()))
d_count = {}
for x in lst:
    d_count[x] = d_count.get(x, 0) + 1
print(d_count)
```

```python
# W6A9: Nhập vào hai dictionary
# Sau đó gộp thành một dictionary mới
# Nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có trong một dict thì giữ nguyên
# In ra dictionary mới theo thứ tự từ điển tăng dần của key

def dict_creation():
    items = input().strip().split()
    d = {}
    for i in range(0, len(items), 2):
        key = items[i]
        value = int(items[i+1])
        d[key] = value
    return d


d1 = dict_creation()
d2 = dict_creation()

d = d1.copy()
for (key, value) in d2.items():
    d[key] = d.get(key, 0) + value

d_sorted = dict(sorted(d.items()))
print(d_sorted)
# -- d.items() trả về list các tuple (key, value)
# -- sorted(d.items()) -> sắp xếp các tuple đó key phần từ đầu (là theo key)
# -- dict() -> chuyển các tuple đó lại thành list

# Sắp xếp theo value
# d_sorted_values = dict(sorted(d.items(), key = lambda x : x[1]))

# Sắp xếp theo value rồi theo theo key
# d_sorted_values_keys = dict(sorted(d.items(), key = lambda x : (x[1], x[0])))

```

```python
# W6A10: Nhập vào list số nguyên và số nguyên dương k.
# In ra list gồm các cặp số khác nhau có tổng bằng k.
# Mỗi cặp là một tuple, hai số trong tuple có thể giống nhau.
# Thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai.
# Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple

lst = list(map(int, input().strip().split()))
k = int(input())
result = []

for i in range(len(lst)):
    for j in range(i+1, len(lst)):
        if lst[i] + lst[j] == k:
            result.append((lst[i], lst[j]))
            
result = list(set(result))
result.sort(key = lambda x: x[0])
print(result)
```

```python
# W6A11: Nhập vào một tuple chứa các số nguyên.
# Hãy tạo ra hai tuple mới: một chứa tất cả các số chẵn và một chữa tất cả các số lẻ từ tuple ban đầu, giữ nguyên thứ tự tương đối
# In ra hai tuple này trên hai dòng riêng biệt.

tpl = tuple(map(int, input().strip().split()))
tpl_even = tuple(x for x in tpl if x % 2 == 0)
tpl_odd = tuple(x for x in tpl if x % 2 != 0)

print(tpl_even)
print(tpl_odd)
```

```python
# W6A12: Nhập vào một list các số nguyên.
# Hãy tìm số xuất hiện nhiều lần nhất trong list.
# Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, in ra số nhỏ nhất trong số đó

lst = list(map(int, input().strip().split()))
d_count = {}
for x in lst:
    d_count[x] = d_count.get(x, 0) + 1

d = sorted(d_count.items(), key = lambda x: (-x[1], x[0]))
# -- Sắp xếp x[1] giảm dần, sau đó sắp xếp x[0] tăng dần
print(d)
print(d[0][0])
             
```

```python
# W6A13: Nhập vào một dictionary có value là duy nhất (không trùng nhau)
# Hãy tạo một dictionary mới bằng các đảo ngược key và value của dictionary ban đầu

lst = list(map(str, input().strip().split()))
d = {}
for i in range(0, len(lst), 2):
    key = lst[i]
    value = lst[i+1]
    d[key] = value

d_new = {}
for (v, k) in d.items():
    d_new[k] = v

print(d)
print(d_new)
```

```python
# W6A14: Nhập 2 list số nguyên, tìm ra các phần tử chung của 2 list và in ra

lst1 = list(map(int, input().strip().split()))
lst2 = list(map(int, input().strip().split()))
set1 = set(lst1)
set2 = set(lst2)
common = []
for x in set1:
    if x in set2 and x not in common:
        common.append(x)

print(common)
```

```python
# W6A15: Nhập vào dictionary, key là chuỗi, value là số nguyên, và một số nguyên k
# Hãy tạo một dict mới chỉ chứa các cặp key-value từ dict ban đầu mà có value lớn hơn k

lst = list(map(str, input().strip().split()))
k = int(input())
d = {}
for i in range(0, len(lst), 2):
    key = lst[i]
    value = int(lst[i+1])
    d[key] = value

d_new = {}
for (key, value) in d.items():
    if value > k:
        d_new[key] = value
print(d_new)
```
