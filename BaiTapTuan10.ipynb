```python
# W9A1: Viết hàm merge hai danh sách đã sắp xếp mà không dùng bất cứ thuật toán sắp xếp nào

def merge(A, B):
    res = []
    a = b = 0
    while a < len(A) and b < len(B):
        if A[a] < B[b]:
            res.append(A[a])
            a += 1
        else:
            res.append(B[b])
            b += 1
    res.extend(A[a:])
    res.extend(B[b:])
    return res
    
A = list(map(int, input().strip().split()))
B = list(map(int, input().strip().split()))

print(*merge(A, B))
```
