```python
# W7A16: 
''' 
Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu mỗi phần tử x trong mảng xuất hiện đúng x lần. 
Kiểm tra xem mảng có phải là mảng đẹp hay không. 
Nếu không hãy xóa đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp.
'''

def count_ocurr(arr):
    dict_count = {}
    for x in arr:
        dict_count[x] = dict_count.get(x,0) + 1
    return dict_count

def pretty_array(dict_count):
    ans = 'YES'
    delete = 0
    for k, v in dict_count.items():
        if k != v:
            ans = 'NO'
            if k > v:
                delete += v
            else: #k < v
                delete += v - k
    if delete == 0:
        return ans
    return ans + ' ' + str(delete)
            

arr = list(map(int, input().strip().split()))
dict_cnt = count_ocurr(arr)
print(pretty_array(dict_cnt))
```

```python
# W7A17:
'''
Tìm phần tử xuất hiện nhiều hơn n/2 lần trong mảng (nếu có) với độ phức tạp thời gian là O(n), độ phức tạp bộ nhớ là O(1), - Boyer–Moore.
Nếu không có phần tử nào thỏa mãn thì không in gì
'''

def find_majority(arr):
    n = len(arr)

    # Find candidate
    candidate = None
    cnt = 0
    for x in arr:
        if cnt == 0:
            candidate = x
            cnt = 1
        elif x == candidate:
            cnt += 1
        else:
            cnt -= 1

    # Check candidate
    if candidate is not None:
        freq = 0
        for x in arr:
            if x == candidate:
                freq += 1
        if freq > (n // 2):
            return candidate

arr = list(map(int, input().strip().split()))
find_majority(arr)
```

```python
# W7A18:
'''
Sắp xếp có điều kiện: Số chẵn đứng trước, số lẻ đứng sau, chẵn tăng dần, lẻ giảm dần
'''

arr = list(map(int, input().strip().split()))
even = []
odd = []
for x in arr:
    if x % 2 == 0:
        even.append(x)
    else:
        odd.append(x)

even.sort()
odd.sort(reverse = True)

result = even + odd

print(result)
```

```python
# W7A19:
'''
Merge 2 mảng đã sắp xếp (có thể tăng dần, có thể giảm dần) mà không dùng bất cứ thuật toán sắp xếp nào
thành một mảng cũng có cùng thứ tự sắp xếp như 2 mảng ban đầu
'''

A = list(map(int, input().strip().split()))
B = list(map(int, input().strip().split()))

# Kiểm tra là tăng dần hay giảm dần
decre = False   # Đang giả sử là dãy tăng
if len(A) > 1 and A[0] != A[-1]:     # Đảm bảo A có nhiều hơn 1 phần tử, và không phải toàn phần tử giống nhau
    decre = A[0] > A[-1]
elif len(B) > 1 and B[0] != B[-1]:   # Trường hợp A chỉ có 1 phần tử hoặc toàn phần tử giống nhau
    decre = B[0] > B[-1]

# Gộp hai mảng
result = []
a = b = 0
while a < len(A) and b < len(B):
    if (not decre and A[a] < B[b]) or (decre and A[a] > B[b]):
        result.append(A[a])
        a += 1
    else:
        result.append(B[b])
        b += 1
result.extend(A[a:])
result.extend(B[b:])
 
print(*result)
```

```python
# W7A21:
'''
Cho N là ngày, T là tháng năm 2020.
Tính xem đó là ngày thứ bao nhiêu của năm đó nếu đó là ngày hợp lệ
'''

def day_th(N, T):
    
    # Check day
    match T:
        case 1|3|5|7|8|10|12:
            max_day = 31
        case 4|6|9|11:
            max_day = 30
        case 2:
            max_day = 29
        case _:
            max_day = -1
    if not (0 < N <= max_day):
        return 'invalid'
    else:
        dayofmonth = [0, 31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        return N + sum(dayofmonth[:T])

N, T = map(int, input().split())
print(day_th(N, T))
        
```

```python
# W7A23:
'''
Đếm số lượng tên trong xâu kí tự.
Tên được định nghĩa là xâu có độ dài >=2 và có chữ cái đầu viết hoa, các chữ cái khác viết thường, chỉ có kí tự cái.
'''

stc = list(input().split())
cnt = 0
for w in stc:
    while len(w) >= 2 and not w[-1].isalnum():
        w = w[:-1]
    if (len(w) >= 2) and (w.isalpha()) and (w[0].isupper()) and (w[1:].islower()):
        cnt += 1
print(cnt)
```

```python
# W7A24:
'''
Tìm số có đúng k chữ số lớn nhất có thể trong một số có n chữ số (coi đầu vào là một số rất lớn - str).
Không được hoán đổi vị trí các chữ số
'''

'''
- Sử dụng stack
- Duyệt từ trái qua phải cho vào stack, nếu chữ số đấy > stack[-1] thì xoá cho đến khi >= thì thôi
Lưu ý số lần xoá (chỉ được xoá len(s) - k lần thôi)
- Lấy k chữ số đầu tiên của stack
'''

s = input()
k = int(input())

stack = []
remove = len(s) - k  # Lượng chữ số phải xoá

for ch in s:
    while (stack) and (remove > 0) and (ch > stack[-1]):
        stack.pop()
        remove -= 1
    stack.append(ch)

result = ''.join(stack[:k])
print(result)
```

```python
# W7A27:
'''
Kiểm tra dãy ngoặc có hợp lệ không
'''

def isvalid_brackets(s):
    brackets = {')':'(',
                ']':'[',
                '}':'{',
                '>':'<' }
    open_brackets = []
    for ch in s:
        if ch in bracket.keys(): # Ngoặc đóng
            if (not open_brackets) or (brackets[ch] != open_brackets.pop()):
                return False
        elif ch in bracket.values(): # Ngoặc mở
            open_brackets.append(ch)

    return not open_brackets

s = input()
print(isvalid_brackets(s))
```

```python
# W7A28:
'''
Tìm dãy các hợp số dài nhất liên tiếp chứa số nguyên < N.
Nếu có nhiều dãy cùng độ dài thì lấy dãy nhỏ nhất (gặp đầu tiên từ trái qua)
'''

def composite_numbers(N):
    # Hàm lọc các hợp số
    composite = [False] * (N+1)
    composite[0] = composite[1] = False
    for i in range(2, int(N**0.5) + 1):
        if not composite[i]:
            for j in range(i**2, N+1, i):
                composite[j] = True
    return composite

N = int(input())

# Các hợp số < N
composite = composite_numbers(N-1)

max_len = 0
cur_len = 0
end_pos = 0

for i in range(N):
    if composite[i]:
        cur_len += 1
        if cur_len > max_len:
            max_len = cur_len
            end_pos = i
    else:
        cur_len = 0

result = list(range(end_pos - max_len + 1, end_pos + 1))
print(*result)
```

```python
# W7A29:
'''
Tìm dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ bàn phím.
Lấy dãy đầu tiên nếu có nhiều dãy có cũng độ dài
'''

arr = list(map(int, input().strip().split()))
max_len = cur_len = 1
end_pos = 0
for i in range(1, len(arr)):
    if arr[i] > arr[i-1]:
        cur_len += 1
        if cur_len > max_len:
            max_len = cur_len
            end_pos = i
    else:
        cur_len = 1

result = arr[end_pos - max_len + 1:end_pos+1]
print(*result)
```

```python
# W7A30: LONGEST INCREASING SUBSEQUENCE O(nlogn)
'''
Tìm dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập vào từ bàn phím.
'''

def binary_search(arr, target):
    # Tìm vị trí đầu tiên >= target
    left, right = 0, len(arr)
    while left < right:
        mid = (left + right) // 2
        if arr[mid] < target:
            left = mid + 1
        else:
            right = mid
    return left

arr = list(map(int, input().strip().split()))
tail = [arr[0]]
for i in range(1, len(arr)):
    if arr[i] > tail[-1]:
        tail.append(arr[i])
    else:
        pos = binary_search(tail, arr[i])
        tail[pos] = arr[i]
        
print(len(tail))
```

```python
# W7A31:
'''
Đổi chỗ đúng 2 chữ số để được số nhỏ nhất có thể
Ý tưởng: Đổi chỗ cho số bé hơn nhất ở xa nhất (nếu có) của số đang xét
'''

def changes(num):
    n = len(num)
    for i in range(n):
        start = i
        min_num = num[i]
        end = None
        for j in range(n-1, i, -1):
            if num[j] < num[i] and num[j] < min_num:
                min_num = num[j]
                end = j
        if end is not None:
            return num[:start] + num[end] + num[start+1:end] + num[start] + num[end+1:]

num = input()
res = changes(num)

if res:
    while res[0] == '0':
        res = res[1:]
else:
    res = num[:len(num)-2] + num[-1] + num[-2]

print(res)
```
