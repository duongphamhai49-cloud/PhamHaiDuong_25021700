```python
# W6A16: Nhập vào mảng n hàng, m cột số nguyên.
# In lại mảng ra màn hình như ma trận hai , định dạng mỗi phần tử 4 vị trí căn phải.

n, m = map(int, input().split())
a = [list(map(int, input().split())) for _ in range(n)]

for row in a:
    print(' '.join(f"{x:4d}" for x in row))
    # -- f"{x:4d}" định dạng số nguyên chiếm đúng 4 kí tự, căn phải
    # 5   -> '   5'
    # 23  -> '  23'
    # 456 -> ' 456'
```

```python
# W6A17: Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và đường chéo phụ

n = int(input())
a = [list(map(int, input().strip().split())) for _ in range(n)]
cheochinh = []
cheophu = []
for i in range(n):
    cheochinh.append(a[i][i])
    cheophu.append(a[n-i-1][i])
print(cheochinh)
print(cheophu)
```

```python
# W6A18: Nhập vào mảng n hàng, m cột và số nguyên k (n, m, k nhập trên cùng hàng)
# Tính và in ra tổng các phần tử trên cột k và hàng k

n, m, k = map(int, input().strip().split())
a = [list(map(int, input().strip().split())) for _ in range(n)]
col_k = 0
row_k = sum(a[k-1])
for i in range(n):
    col_k += a[i][k-1]
print(col_k)
print(row_k)
```

```python
# W6A19: Viết hàm đưa vào một list số nguyên, tìm và trả về vị trí có giá trị lớn nhất trong list

def index_of_max(a):
    max_value = max(a)
    for i in range(len(a)):
        if a[i] == max_value:
            return i

a = list(map(int, input().strip().split()))
print(index_of_max(a))
```

```python
# W6A20: Viết hàm đưa vào một list số nguyên và số nguyên dương k.
# Tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list
# Không có thì trả về -1

def index_of_k(a,k):
    for i in range(len(a)):
        if a[i] == k:
            return i
    return -1

a = list(map(int, input().strip().split()))
k = int(input())
print(index_of_k(a,k))
```

```python
# W6A21: Nhập một dòng gồm các chuỗi.
# Tạo dictionary ánh xạ chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

lst = list(map(str, input().strip().split()))
d = {}
for i, key in enumerate(lst):
    if key not in d:
        d[key] = [i]
    else:
        d[key].append(i)
        
for x in d:
    d[x] = tuple(d[x])

print(d)
```

```python
# W6A23: Viết hàm xóa đi các số <x có trong mảng

def remove_elements(a, x):
    return [i for i in a if i >= x]

lst = list(map(int, input().strip().split()))
x = int(input())
print(remove_elements(lst, x))
```

```python
# W6A24: Viết hàm đẩy tất cả các số 0 về đầu (về cuối), các số còn lại vẫn giữa nguyên vị trí trước sau

def push_zeros_head(a):
    for i in range(len(a)):
        if a[i] == 0:
            a.pop(i)
            a.insert(0, 0)
    return a

def push_zeros_back(a):
    for i in range(len(a)-1,-1,-1):
        if a[i] == 0:
            a.pop(i)
            a.append(0)
    return a
    
lst = list(map(int, input().strip().split()))
print(push_zeros_head(lst))
print(push_zeros_back(lst))
```

```python
# W6A25: Cho dãy chỉ gồm N giá trị từ 0 đến 9.
# Kiểm tra xem giá trị nào không có trong mảng.

lst = list(map(int, input().strip().split()))
cnt = [0]*10
for i in lst:
    cnt[i] += 1

for i in range(len(cnt)):
    if cnt[i] == 0:
        print(i, end=' ')
```

```python
# W6A26: Tính tổng các phần tử cực đại địa phương
# Không tính phần tử đầu và cuối

lst = list(map(int, input().strip().split()))
s = 0
for i in range(1, len(lst)-1):
    if lst[i-1] < lst[i] and lst[i+1] < lst[i]:
        s += lst[i]

print(s)
```

```python
# W6A27: Cho một dãy số nguyên. Tìm và in ra vị trí các số < x
# In ra -1 nếu không có

lst = list(map(int, input().strip().split()))
x = int(input())
if x <= min(lst):
    print(-1)
else:
    for i in range(len(lst)):
        if lst[i] < x:
            print(i, end=' ')
```

```python
# W6A28: Cho một dãy số nguyên, kiểm tra xem nó là dãy tăng / giảm / vô danh

lst = list(map(int, input().strip().split()))

increase = True
decrease = True

for i in range(1, len(lst)):
    if lst[i] < lst[i-1]:
        increase = False
    if lst[i] > lst[i-1]:
        decrease = False

if increase:
    print('Tăng')
elif decrease:
    print('Giảm')
else:
    print('Vô danh')
```
